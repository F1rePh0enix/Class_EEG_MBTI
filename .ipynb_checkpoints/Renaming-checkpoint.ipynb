```python
import json

# UPPERCASE mapping
code_map = {
    "CENTER_BACK": "CB",
    "CENTER_FRONT": "CF",
    "BACK": "BB",
    "LEFT": "LL",
    "BACK_LEFT": "BL",
    "FRONT": "FF",
    "RIGHT": "RR",
    "OPEN_ANALYTIC": "OO",
    "HALO": "HH",
    "STRONG_STARBURST": "SU",
    "HIDDEN_STARBURST": "DU",
    "WEAK_STARBURST": "WU",
    "EVEN_FIELD": "EE",
    "DIAMONDS": "DD",
    "AUDIO_KINESTHETIC": "AK",
    "SUPER_AUDIO": "AA",
    "RANDOM_FIELD": "NN",
    "FOCUSED_HOLISTIC": "SS",
    "PILLAR": "PP",
    "ZIGZAG": "ZZ"
}

def encode_top_patterns(row):
    if not isinstance(row, str):
        return None
    try:
        parsed = json.loads(row)
        if not isinstance(parsed, list):
            return None
        codes = [code_map.get(p.upper()) for p in parsed if p.upper() in code_map]
        if not codes:
            return None
        return ' '.join(sorted(codes))
    except json.JSONDecodeError:
        return None

# Apply and create a new column
db['top_patterns_encoded'] = db['top_patterns'].apply(encode_top_patterns)
display()
```

```python
print(repr(db['top_skills'].iloc[0]))
```

```python

def clean_top_skills(row):
    if not isinstance(row, str):
        return None
    try:
        parsed = json.loads(row)
        if not isinstance(parsed, list):
            return None
        cleaned = sorted(tag.strip().upper() for tag in parsed if isinstance(tag, str))
        if not cleaned:
            return None
        return ' '.join(cleaned)
    except json.JSONDecodeError:
        return None

# Create new readable column
db['top_skills_cleaned'] = db['top_skills'].apply(clean_top_skills)
display(db)
```
